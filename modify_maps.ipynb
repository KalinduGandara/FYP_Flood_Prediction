```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import rasterio as rio
import os
```

```python
input_dir = 'data\kaduwela_8\DEM5m_2D'
output_dir = 'data\kaduwela_8\DEM5m_2D_new'

for filename in os.listdir(input_dir):
    if filename.endswith('.wd'):
        with rio.open(os.path.join(input_dir, filename)) as src:
            data = src.read()
            data = data / 6
            data = np.round(data, 3)

            profile = src.profile
            with rio.open(os.path.join(output_dir, filename), 'w', **profile) as dst:
                dst.write(data)
```

```python
tif_file = 'data\kaduwela_8\DEM5m_2D\Run2-0120.wd'

with rio.open(tif_file) as src:
    img = src.read(1)
    plt.imshow(img)
    plt.colorbar()
    plt.show()
```
